# 2. Ventas por categoría

```python
# Leer los datos de cada tienda
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Eliminar espacios adicionales en los nombres de las columnas
tienda.columns = tienda.columns.str.strip()
tienda2.columns = tienda2.columns.str.strip()
tienda3.columns = tienda3.columns.str.strip()
tienda4.columns = tienda4.columns.str.strip()

# Función para calcular la cantidad de productos vendidos por categoría
def calcular_categorias_vendidas(df, tienda_name):
    # Agrupar los datos por la categoría y contar el número de productos vendidos
    categorias_vendidas = df.groupby('Categoría del Producto')['Precio'].count().sort_values(ascending=False)

    # Mostrar el resultado
    print(f"Categorías más populares - {tienda_name}:")
    print(categorias_vendidas)
    print("\n")

# Calcular y mostrar para cada tienda
calcular_categorias_vendidas(tienda, "Tienda 1")
calcular_categorias_vendidas(tienda2, "Tienda 2")
calcular_categorias_vendidas(tienda3, "Tienda 3")
calcular_categorias_vendidas(tienda4, "Tienda 4")
```

```python
# Obtener las categorías vendidas para cada tienda
categorias_tienda1 = calcular_categorias_vendidas(tienda)
categorias_tienda2 = calcular_categorias_vendidas(tienda2)
categorias_tienda3 = calcular_categorias_vendidas(tienda3)
categorias_tienda4 = calcular_categorias_vendidas(tienda4)

# Crear un DataFrame para combinar todas las tiendas
df_completo = pd.DataFrame({
    'Tienda 1': categorias_tienda1,
    'Tienda 2': categorias_tienda2,
    'Tienda 3': categorias_tienda3,
    'Tienda 4': categorias_tienda4
}).fillna(0)  # Rellenar los valores faltantes con 0 para aquellas categorías no presentes en todas las tiendas

# Crear el gráfico de barras agrupadas
df_completo.plot(kind='bar', figsize=(12, 6), color=['lightblue', 'lightgreen', 'lightcoral', 'lightyellow'], edgecolor='black')

# Personalizar el gráfico
plt.title("Cantidad de Productos Vendidos por Categoría en Cada Tienda", fontsize=16)
plt.xlabel("Categoría del Producto", fontsize=12)
plt.ylabel("Cantidad de Productos Vendidos", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Calcular los productos más y menos vendidos por tienda
def productos_mas_menos_vendidos(tienda, tienda_nombre):
    # Contar el número de ventas por producto
    productos_ventas = tienda.groupby('Producto')['Cantidad de cuotas'].sum().reset_index()

    # Ordenar los productos de mayor a menor cantidad de ventas
    productos_ventas = productos_ventas.sort_values(by='Cantidad de cuotas', ascending=False)

    # Productos más vendidos
    productos_mas_vendidos = productos_ventas.head(5)  # Top 5 productos más vendidos

    # Productos menos vendidos (con al menos una venta)
    productos_menos_vendidos = productos_ventas.tail(5)  # Top 5 productos menos vendidos

    print(f"\n{tienda_nombre} - Productos más vendidos:")
    print(productos_mas_vendidos[['Producto', 'Cantidad de cuotas']])

    print(f"\n{tienda_nombre} - Productos menos vendidos:")
    print(productos_menos_vendidos[['Producto', 'Cantidad de cuotas']])

# Aplicar la función a cada tienda
productos_mas_menos_vendidos(tienda, "Tienda 1")
productos_mas_menos_vendidos(tienda2, "Tienda 2")
productos_mas_menos_vendidos(tienda3, "Tienda 3")
productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Función para obtener los productos más vendidos y menos vendidos por tienda
def obtener_productos_por_tienda(tienda, tienda_nombre):
    # Agrupar los datos por 'Producto' y sumar la cantidad de ventas
    productos_ventas = tienda.groupby('Producto')['Cantidad de cuotas'].sum().reset_index()
    # Ordenar por la cantidad de ventas en orden descendente
    productos_ventas = productos_ventas.sort_values(by='Cantidad de cuotas', ascending=False)

    # Obtener los 5 productos más vendidos
    productos_mas_vendidos = productos_ventas.head(5)
    productos_mas_vendidos['Tienda'] = tienda_nombre

    # Obtener los 5 productos menos vendidos
    productos_menos_vendidos = productos_ventas.tail(5)
    productos_menos_vendidos['Tienda'] = tienda_nombre

    return productos_mas_vendidos, productos_menos_vendidos

# Obtener los productos más y menos vendidos para cada tienda
tienda_1_mas_vendidos, tienda_1_menos_vendidos = obtener_productos_por_tienda(tienda, "Tienda 1")
tienda_2_mas_vendidos, tienda_2_menos_vendidos = obtener_productos_por_tienda(tienda2, "Tienda 2")
tienda_3_mas_vendidos, tienda_3_menos_vendidos = obtener_productos_por_tienda(tienda3, "Tienda 3")
tienda_4_mas_vendidos, tienda_4_menos_vendidos = obtener_productos_por_tienda(tienda4, "Tienda 4")

# Concatenar los datos de las 4 tiendas para los más vendidos
productos_mas_vendidos = pd.concat([tienda_1_mas_vendidos, tienda_2_mas_vendidos, tienda_3_mas_vendidos, tienda_4_mas_vendidos])

# Concatenar los datos de las 4 tiendas para los menos vendidos
productos_menos_vendidos = pd.concat([tienda_1_menos_vendidos, tienda_2_menos_vendidos, tienda_3_menos_vendidos, tienda_4_menos_vendidos])

# Crear los gráficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))

# Gráfico para los productos más vendidos
sns.barplot(x='Producto', y='Cantidad de cuotas', hue='Tienda', data=productos_mas_vendidos, ax=ax1)
ax1.set_title('5 Productos Más Vendidos por Tienda')
ax1.set_xlabel('Producto')
ax1.set_ylabel('Cantidad de Ventas')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')  # Gira las etiquetas del eje X

# Gráfico para los productos menos vendidos
sns.barplot(x='Producto', y='Cantidad de cuotas', hue='Tienda', data=productos_menos_vendidos, ax=ax2)
ax2.set_title('5 Productos Menos Vendidos por Tienda')
ax2.set_xlabel('Producto')
ax2.set_ylabel('Cantidad de Ventas')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')  # Gira las etiquetas del eje X

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Definimos una función para calcular el promedio de costo de envío por tienda
def calcular_promedio_costo_envio(tienda):
    # Calculamos el promedio de la columna 'Costo de envío' y lo redondeamos a 2 decimales
    costo_envio_promedio = round(tienda['Costo de envío'].mean(), 2)
    return costo_envio_promedio

# Calculamos el promedio de costo de envío para cada tienda
tienda_1_promedio_costo_envio = calcular_promedio_costo_envio(tienda)
tienda_2_promedio_costo_envio = calcular_promedio_costo_envio(tienda2)
tienda_3_promedio_costo_envio = calcular_promedio_costo_envio(tienda3)
tienda_4_promedio_costo_envio = calcular_promedio_costo_envio(tienda4)

# Mostrar los resultados por tienda
print("Promedio de Costo de Envío para Tienda 1:", tienda_1_promedio_costo_envio)
print("Promedio de Costo de Envío para Tienda 2:", tienda_2_promedio_costo_envio)
print("Promedio de Costo de Envío para Tienda 3:", tienda_3_promedio_costo_envio)
print("Promedio de Costo de Envío para Tienda 4:", tienda_4_promedio_costo_envio)
```
